```python
import math
import time
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy import linalg as LA
```

```python
circle = np.loadtxt(open("circle.txt", "rb"), delimiter=",")
moon = np.loadtxt(open("moon.txt", "rb"), delimiter=",")
```

```python
def rand_float(a, b):
    return random.uniform(a, b)

def pts_dis(A, B):
    return math.hypot(A[0] - B[0], A[1] - B[1])

def plot_cluster(D, pred_c, mu, c):
    clr = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']
#     mkr = ['o', 'x', 'v', '^']
    x = D[pred_c==c][:, 0]
    y = D[pred_c==c][:, 1]
    plt.scatter(x, y, color=clr[c], alpha=0.3, marker='.')
    if np.all(mu) != None:
        plt.scatter(mu[c][0], mu[c][1], color=clr[c], marker='*', s=100)
    
def rbf_kernel(a, b, gamma):
    return math.exp(-gamma * ((pts_dis(a, b))**2))

def gram_matrix(datas, gamma):
    gram = np.zeros((len(datas), len(datas)))
    for p in range(len(datas)):
        for q in range(len(datas)):
            gram[p][q] = rbf_kernel(datas[p], datas[q], gamma)
    return gram

def cluster_count(c, k):
    return np.count_nonzero(c == k)
```

# Spectral Clustering

```python
def build_similarity(datas, gamma):
    similarity = np.zeros((len(datas), len(datas)))
    for i in range(len(datas)):
        for j in range(i, len(datas)):
            similarity[i][j] = rbf_kernel(datas[i], datas[j], gamma)
    for i in range(1, len(datas)):
        for j in range(i):
            similarity[i][j] = similarity[j][i]
    return similarity

def build_degree(datas, similarity):
    degree = np.zeros((len(datas), len(datas)))
    for i in range(len(datas)):
            degree[i][i] = np.sum(similarity[i])
    return degree

def build_laplacian(similarity, degree):
    return degree - similarity

def find_eigengap(e_val):
    gap = np.zeros(len(e_val))
    for i in range(1, len(e_val)):
        gap[i] = abs(e_val[i] - e_val[i-1])
    
    print("e_val : ", e_val)
    if abs(e_val[0]) <= 1e-8:
        print("first == 0")
        gap_idx = np.argmax(gap[2:])+2
    else:
        gap_idx = np.argmax(gap)
    return gap_idx
```

```python
def Spectral_Clustering(datas, k, gamma):
    similarity = build_similarity(datas, gamma)
    degree = build_degree(datas, similarity)
    laplacian = build_laplacian(similarity, degree)
    
    e_val, e_vec = LA.eig(laplacian)
#     print(e_val == e_val.sort())
#     gap_idx = find_eigengap(e_val)
```
